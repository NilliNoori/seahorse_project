## Genre driven consumption insights

Provide a comparative analysis of spend-per-head and product mix across music genres, surfaced in a simple dashboard that lets the programming team filter by season or venue format.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# Dummy data for genre-driven spend-per-head and product mix
genres = ['Techno', 'Hip-Hop', 'House', 'Pop', 'R&B']
spend_per_head = [18, 15, 17, 14, 13]  # Euros
beer_percentage = [40, 30, 35, 45, 25]  # % of total sales
cocktail_percentage = [30, 35, 25, 20, 40]
soft_drink_percentage = [30, 35, 40, 35, 35]

# Create a figure with two subplots: bar chart for spend-per-head, stacked bar for product mix
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

# Bar chart: Spend per head by genre
ax1.bar(genres, spend_per_head, color='gray')
ax1.set_title('Spend per Head by Music Genre')
ax1.set_ylabel(' per person')
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# Stacked bar chart: Product mix by genre
bar1 = ax2.bar(genres, beer_percentage, label='Beer')
bar2 = ax2.bar(genres, cocktail_percentage, bottom=beer_percentage, label='Cocktails')
bottom2 = [beer + cocktail for beer, cocktail in zip(beer_percentage, cocktail_percentage)]
bar3 = ax2.bar(genres, soft_drink_percentage, bottom=bottom2, label='Soft Drinks')

ax2.set_title('Product Mix by Music Genre')
ax2.set_ylabel('% of Total Sales')
ax2.legend()
ax2.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## Bar Revenue Trends by Ticket Category

Build a lightweight simulation tool that tests different early-bird ticket quantities & prices, showing the impact on both ticket revenue and bar spend based on historic arrival-time vs-spend patterns.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Dummy data
ticket_types = ['Early Bird', 'Regular', 'Late']
drinks_sold = [1200, 2000, 800]
tickets_sold = [300, 500, 200]

# Create a DataFrame for easier plotting
df = pd.DataFrame({
    'Ticket Type': ticket_types,
    'Tickets Sold': tickets_sold,
    'Drinks Sold': drinks_sold
})

# Plot: Drinks sold vs Tickets sold per ticket type
fig, ax = plt.subplots()
bar_width = 0.35
index = np.arange(len(df))

ax.bar(index, df['Tickets Sold'], bar_width, label='Tickets Sold')
ax.bar(index + bar_width, df['Drinks Sold'], bar_width, label='Drinks Sold')

ax.set_xlabel('Ticket Type')
ax.set_ylabel('Quantity')
ax.set_title('Tickets Sold vs. Drinks Sold by Ticket Type')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(df['Ticket Type'])
ax.legend()
plt.tight_layout()

plt.show()
```
